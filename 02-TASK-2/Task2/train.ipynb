```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
import pickle
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('train.csv')
```

```python
df_test=pd.read_csv('test.csv')
```

```python
df.info()
```

```python
df_test.info()
```

```python
df.drop('Name', axis=1,inplace=True)
```

```python
df.drop('Cabin', axis=1,inplace=True)
```

```python
df_test.drop('Name', axis=1,inplace=True)
```

```python
df_test.drop('Cabin', axis=1,inplace=True)
```

```python
fill_NA_obj=['HomePlanet','CryoSleep','Destination','VIP']
for i in fill_NA_obj:
    df[i]=df[i].fillna(df[i].mode()[0])
```

```python
fill_NA_obj = ['HomePlanet', 'CryoSleep', 'Destination', 'VIP']
for i in fill_NA_obj:
    df_test[i] = df_test[i].fillna(df_test[i].mode()[0])
```

```python
fill_NA_flt=['Age','RoomService','FoodCourt','ShoppingMall','Spa','VRDeck']
for i in fill_NA_flt:
    df[i]=df[i].fillna(df[i].median())
```

```python
fill_NA_flt = ['Age', 'RoomService', 'FoodCourt', 'ShoppingMall', 'Spa', 'VRDeck']
for i in fill_NA_flt:
    df_test[i] = df_test[i].fillna(df_test[i].median())
```

```python
obj_to_int=['HomePlanet','Destination']
for i in obj_to_int:
    label_encoder=LabelEncoder()
    df[i]=label_encoder.fit_transform(df[i])
```

```python
obj_to_int = ['HomePlanet', 'Destination']
for i in obj_to_int:
    label_encoder = LabelEncoder()
    df_test[i] = label_encoder.fit_transform(df_test[i])
```

```python
float_to_int=['Age','RoomService','FoodCourt','ShoppingMall','Spa','VRDeck']
for i in float_to_int:
    df[i]=df[i].astype('int64')
```

```python
float_to_int = ['Age', 'RoomService', 'FoodCourt', 'ShoppingMall', 'Spa', 'VRDeck']
for i in float_to_int:
    df_test[i]=df_test[i].astype('int64')
```

```python
bool_to_int=['CryoSleep','VIP','Transported']
for i in bool_to_int:
    df[i]=df[i].astype('int64')
```

```python
bool_to_int = ['CryoSleep', 'VIP']
for i in bool_to_int:
    df_test[i]=df_test[i].astype('int64')
```

```python
df.to_csv('train.csv',index=False)
```

```python
df_test.to_csv('test.csv',index=False)
```

```python
df['PassengerId'].nunique()
```

```python
df.drop('PassengerId',axis=1,inplace=True)
```

```python
df.info()
```

```python
train_x=df.drop(columns='Transported')
train_y=df['Transported']
```

```python
test_x=df_test.drop(columns='PassengerId')
test_id=df_test['PassengerId']
```

```python
model_RF=RandomForestRegressor(n_estimators=100 , random_state=42)
model_RF.fit(train_x,train_y)
```

```python
model_LR = LinearRegression()
model_LR.fit(train_x, train_y)
```

```python
pickle.dump(model_RF, open('pickle_RF.pkl','wb'))
```

```python
pickle.dump(model_LR, open('pickle_LR.pkl','wb'))
```

```python
model_prediction=model_RF.predict(test_x)
```

```python
model_prediction=model_LR.predict(test_x)
```

```python
submission = pd.DataFrame({'PassengerId': test_id, 'Transported': model_prediction}) 

submission.to_csv('submission_RF.csv', index=False)
```
